```python
%matplotlib inline
import numpy as np
import numpy.random as rnd
import scipy.misc as misc
import matplotlib.pyplot as plt
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
plt.rcParams["figure.figsize"] = (5 * (1 + np.sqrt(5)) / 2, 5)
plt.rcParams["lines.linewidth"] = 2.5
plt.rcParams["ytick.labelsize"] = 12
plt.rcParams["xtick.labelsize"] = 12
plt.rcParams["axes.labelsize"] = 20
```

###Defino las distribuciones de probabilidad que necesito usar

```python
poisson = lambda x, μ: np.exp(-μ) * np.power(μ, x) / misc.factorial(x)
binomial = lambda x, N, p: misc.comb(N, x) * p**x * (1 - p)**(N-x)
gaussiana = lambda x, μ, σ: 1/(np.sqrt(2*np.pi)*σ) * np.exp(-1/2*np.power((x - μ)/σ, 2))
```

###Grafico la distribución binomial que se pide, y una gaussiana usando la esperanza y la varianza de la binomial, observando si es válido el teorema central del límite

```python
def G6_E4_a(N = 5, p = 0.2):
    #4σ del la esperanza para una escala razonable
    if N*p - 4*np.sqrt(N*p*(1-p)) < 0:
        x_min = 0
    else:
        x_min = N*p - 4*np.sqrt(N*p*(1-p))
    x_max = N*p + 4*np.sqrt(N*p*(1-p))
    t = np.arange(x_min, x_max, 1.0)
    x = np.linspace(x_min, x_max,1000)
    plt.figure()
    plt.plot(t, binomial(t, N, p), 'bo', label = 'Distribución Binomial')
    plt.plot(x, gaussiana(x, N*p, np.sqrt(N*p*(1-p))),'r-', label = "Distribución Gaussiana")
    plt.legend(loc=0)
    
G6_E4_a(N=5, p = 0.2)
G6_E4_a(N = 30, p = 0.4)
```

###Grafico la distribuciones de Poisson que se piden, y además una distribución gaussiana usando la esperanza y la varianza de la poissoneana, observando si es válido el teorema central del límite

```python
def G6_E4_b(μ = 4):
    #4σ del la esperanza para una escala razonable
    if μ - 4*np.sqrt(μ) < 0:
        x_min = 0
    else:
        x_min = μ - 4*np.sqrt(μ)
    x_max = μ + 4*np.sqrt(μ)
    t = np.arange(x_min, x_max, 1.0)
    x = np.linspace(x_min, x_max,1000)
    plt.figure()
    plt.plot(t, poisson(t, μ), 'bo', label = "Distribución de Poisson")
    plt.plot(x, gaussiana(x, μ, np.sqrt(μ)),'r-', label = "Distribución Gaussiana")
    plt.legend(loc=0)

G6_E4_b(4)
G6_E4_b(10)
G6_E4_b(50)
```
